```python
def should_go_outside():
    print("Start")
    
    # First decision point
    is_raining = input("Is it raining? (yes/no): ").lower()
    if is_raining == "yes":
        # Second decision point
        has_umbrella = input("Do you have an umbrella? (yes/no): ").lower()
        if has_umbrella == "no":
            print("Wait a while.")
            # Re-check rain status after waiting
            is_raining = input("Is it still raining? (yes/no): ").lower()  # This line was missing
            if is_raining == "no":
                print("Go outside.")
            else:
                print("Stay inside.")
        else:
            print("Go outside with your umbrella.")
    else:
        print("Go outside.")

    print("End")

# Run the function
should_go_outside()
```

```python
import random

def number_guessing_game():
    print("Welcome to the Number Guessing Game!")
    print("Choose difficulty:")
    print("1: Easy (1-10, 5 guesses)")
    print("2: Medium (1-20, 6 guesses)")
    print("3: Hard (1-50, 7 guesses)")
    
    # Set difficulty
    while True:
        try:
            difficulty = int(input("Enter 1-3: "))
            if difficulty in (1, 2, 3):
                break
            print("Please enter 1, 2, or 3")
        except ValueError:
            print("Numbers only please!")

    # Configure game based on difficulty
    ranges = {1: (1,10,5), 2: (1,20,6), 3: (1,50,7)}
    min_num, max_num, max_guesses = ranges[difficulty]
    
    secret_number = random.randint(min_num, max_num)
    print(f"\nI'm thinking of a number between {min_num} and {max_num}...")

    # Game logic
    for guesses_taken in range(1, max_guesses + 1):
        print(f"\nGuess {guesses_taken}/{max_guesses}")
        
        while True:
            try:
                guess = int(input("Your guess: "))
                if min_num <= guess <= max_num:
                    break
                print(f"Please enter a number between {min_num}-{max_num}")
            except ValueError:
                print("Numbers only please!")

        # Give hints
        if guess < secret_number:
            print("Too low!", end=' ')
            if secret_number - guess > 15:
                print("(Way too low!)")
            elif secret_number - guess > 5:
                print("(Pretty low)")
        elif guess > secret_number:
            print("Too high!", end=' ')
            if guess - secret_number > 15:
                print("(Way too high!)")
            elif guess - secret_number > 5:
                print("(Pretty high)")
        else:
            break  # Correct guess

    # Win/lose message
    if guess == secret_number:
        print(f"\n🎉 Correct! You guessed it in {guesses_taken} tries!")
    else:
        print(f"\n😢 Game over! The number was {secret_number}")

    # Play again?
    if input("\nPlay again? (y/n): ").lower() == 'y':
        number_guessing_game()
    else:
        print("Thanks for playing!")

# Start the game
number_guessing_game()
```

```python
import random

def rock_paper_scissors():
    choices = ['rock', 'paper', 'scissors']
    score = {'player': 0, 'computer': 0}

    while True:
        print(f"\nScore: Player {score['player']} - Computer {score['computer']}")
        player_choice = input("Choose rock/paper/scissors (or 'quit' to exit): ").lower()

        # Exit condition
        if player_choice == 'quit':
            print("Final Score:", score)
            break

        # Validate input
        if player_choice not in choices:
            print("Invalid choice! Try again.")
            continue

        # Computer's random choice
        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")
        
        # Determine winner
        if player_choice == computer_choice:
            print("It's a tie!")
        elif (
            (player_choice == 'rock' and computer_choice == 'scissors') or
            (player_choice == 'paper' and computer_choice == 'rock') or
            (player_choice == 'scissors' and computer_choice == 'paper')
        ):
            print("You win!")
            score['player'] += 1
        else:
            print("Computer wins!")
            score['computer'] += 1

print("=== Rock-Paper-Scissors Game ===")
rock_paper_scissors()
```
